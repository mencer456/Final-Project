# City Traffic Signs

```python
import pandas as pd
import matplotlib.pyplot as plt

#imports dataset
signs = pd.read_csv("https://data.wprdc.org/datastore/dump/d078a6b5-83a3-4723-a3a9-5371cfe1cc0c")
signs
```

```python
#gets rid of rows with no neighborhood
droppedSigns = signs.dropna(subset = ["neighborhood"], inplace=True) 
signs.sort_values(by=['neighborhood'],ascending=True)
```

# Number of traffic signs in each neighborhood

```python
#organizes the neighborhoods by most traffic signs
signs["neighborhood"].value_counts()
```

```python
#bar graph of top 10 neighborhoods with the most traffic signs
signs1 = signs['neighborhood'].value_counts().sort_values(ascending=True).tail(10)
signs1.plot.barh(x="neighborhood",y="mounting_fixtures")
```

```python
#finding number of each specific sign
signs["mounting_fixture"].value_counts()
```

```python
#testing other methods of plotting datasets
signs["neighborhood"].value_counts().plot.area()
```
